## 1.d Visualization

```python
import plotly.express as px
import re

# Read the frequent patterns from file
with open("frequentPatterns.txt", "r", encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip()]

# Parse and find the longest pattern
def extract_points(pattern_line):
    return re.findall(r"POINT\(([-\d\.]+) ([-\d\.]+)\)", pattern_line)

# Apply parsing and keep the longest pattern
longest = max(lines, key=lambda line: len(extract_points(line)))
points = extract_points(longest)

# Create DataFrame for plotting
df = pd.DataFrame(points, columns=["lon", "lat"]).astype(float)

# Plot using plotly express
fig = px.line_mapbox(df,
                     lat="lat",
                     lon="lon",
                     zoom=5,
                     height=500,
                     title="Longest Frequent Pattern Path",
                     center={"lat": df["lat"].mean(), "lon": df["lon"].mean()})
fig.update_layout(mapbox_style="open-street-map")
fig.show()
```
